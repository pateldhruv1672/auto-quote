```python
import pandas as pd
# Read file line-by-line and split on the first tab to avoid malformed lines with extra tabs
rows = []
with open("fraud_call.file", "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        parts = line.rstrip("\n").split("\t", 1)
        if len(parts) == 2:
            rows.append(parts)
        else:
            # If no tab found, keep the whole line as clue and mark type as unknown
            # (this is defensive; dataset appears to have label + tab + text)
            rows.append([parts[0], ""])
df = pd.DataFrame(rows, columns=["type", "clue"])
print("Loaded dataset shape:", df.shape)
df.head(10)
```
